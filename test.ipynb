# Create patterns

## Create circles

```python
r=1
x=0
y=0
lst = [(0,0)]
while(r < 150):
    for i in range(int(-r), int(r)+1):
        for j in range(int(-r), int(r)+1):
            if (i*i + j*j) <= (r*r) and (i,j) not in lst:
                lst.append((i,j))
    r+=0.2
```

## Create pattern with dens=5

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def delete_point(x0,y0,arr):
    point = [x0-156,y0-156]

    while True:
        while True:
            if point in arr:
                arr.remove(point)
            point[1] += 6

            if point[1] > 150:
                break

        point[0] += 6
        point[1] = y0-156

        if point[0] > 150:
            break
```

```python
def save_pattern(dx,dy,arr):
    coords = np.array([(0, -2), (-1,1), (-2, -3), (-3, -1), (-4, 0), (-5, 2)])
    coords[:, 0] -= (dx - 1)
    coords[:, 1] -= (dy - 1)

    for x,y in coords:
        delete_point(x,y,arr)

    if [0,0] in arr:
        arr.remove([0,0])

    x = np.array(arr[:1000])
    plt.figure(figsize=(10,10))
    plt.scatter(x[:,0], x[:,1])
    plt.scatter(0,0,c='red')
    plt.xlim((-30,30))
    plt.ylim((-30,30))

    with open("algorithms\\build_patterns\\"+ f"new_dens5_{dx}_{dy}.txt", 'w') as f:
        json.dump(arr,f)
```

```python
for dx in range(6):
    for dy in range(6):
        with open("algorithms\\build_patterns\\circle2.txt") as f:
            arr = json.load(f)
        print(dx, dy)
        save_pattern(dx, dy, arr)
```

# Visualization

```python
import matplotlib.pyplot as plt
import os
import json
```

```python
filename = '2024-07-13 16.39.33.json'
with open('turns\\units\\' + filename) as f:
    data = json.load(f)
    units = data

with open('turns\\world\\' + filename) as f:
    data = json.load(f)
    world = data
```

```python
def visualize(units, world, title="turn x", save=None):
    
    our_base = units.get("base", []) or []
    zombies = units.get("zombies", []) or []
    an_player_bases = units.get("enemyBlocks", []) or []
    zpots = world.get("zpots", []) or []

    plt.figure(figsize=(10,10))

    for cell in our_base:
        if "isHead" in cell:
            plt.scatter(cell['x'], cell['y'], c='black', alpha=cell['health']/300, linewidths=20, s=200)
        else:
            plt.scatter(cell['x'], cell['y'], c='blue', alpha=cell['health']/100, linewidths=8, s=40)
    
    for cell in an_player_bases:
        if cell['health'] > 100:
            plt.scatter(cell['x'], cell['y'], c='red', alpha=cell['health']/300, linewidths=10, s=200)
        else:
            plt.scatter(cell['x'], cell['y'], c='red', alpha=cell['health']/100, linewidths=8, s=40)

    for cell in zombies:
        plt.scatter(cell['x'], cell['y'], c='green', s=150)
        plt.text(cell['x']-0.2, cell['y']-0.2, cell['type'][0])
        if (cell['direction'] == 'up'):
            plt.arrow(cell['x'], cell['y'] + 0.3, 0, 0.2)
        elif (cell['direction'] == 'left'):
            plt.arrow(cell['x'] - 0.3, cell['y'], -0.2, 0)
        elif (cell['direction'] == 'right'):
            plt.arrow(cell['x'] + 0.3, cell['y'], 0.2, 0)
        elif (cell['direction'] == 'down'):
            plt.arrow(cell['x'], cell['y'] - 0.3, 0, -0.2)

    for zpot in zpots:
        if zpot["type"] == 'default':
            plt.scatter(zpot['x'], zpot['y'], c='yellow', s=150)
        else:
            plt.scatter(zpot['x'], zpot['y'], c='magenta', s=150)


    plt.title(title)
    if save:
        plt.savefig(os.path.join('turns images', save + '.png'))
        plt.close()
    
```

```python
filename = '2024-07-13 18.42.56.json'
with open('dataset\\final-8\\units\\' + filename) as f:
    data = json.load(f)
    units = data

with open('dataset\\final-8\\world\\' + filename) as f:
    data = json.load(f)
    world = data

visualize(units, world, title='turn')
```
